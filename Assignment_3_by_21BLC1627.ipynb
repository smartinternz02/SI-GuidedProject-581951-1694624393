##Check for Missing values and deal with them.


```python
df.isnull().sum()
```

####Sex is a categorial column will null values which will be filled wiht mode
###Where as rest columns with null value are numerical column, hence we will be using mean to fill them

```python
df.sex.value_counts()
```

```python
# To fill Sex column null space
df.sex = df.sex.fillna('MALE')
df.sex.value_counts()
```

The null spaces have been fill still the '.' dot is left so we will use replace() funtion

```python
df.sex = df.sex.replace('.','MALE')
df.sex.value_counts()
```

```python
print(df.culmen_length_mm.mean())
print(df.culmen_depth_mm.mean())
print(df.flipper_length_mm.mean())
print(df.body_mass_g.mean())
```

```python
#Replacing all values with mean of the column, can also be replaced by medin value
df.culmen_length_mm = df.culmen_length_mm.fillna(df.culmen_length_mm.mean())
df.culmen_depth_mm = df.culmen_depth_mm.fillna(df.culmen_depth_mm.mean())
df.flipper_length_mm = df.flipper_length_mm.fillna(df.flipper_length_mm.mean())
df.body_mass_g = df.body_mass_g.fillna(df.body_mass_g.mean())
#   df=df.fillna(df.mean()) can also use this line of code
df.isnull().sum()
```
